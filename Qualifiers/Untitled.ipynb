```python
best_model= compare_models()
```

```python
et = create_model('et')
```

```python
predictions = predict_model(et,data=test)
```

```python
submission['nerdiness'] = predictions['Label']
```

```python
blended = blend_models(estimator_list = best_2, fold = 5, method = 'soft')
```
